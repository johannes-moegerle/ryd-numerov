# Simple example of integrating the hydrogen wavefunction

```python
from numerov.integrate import integrate_hydrogen
```

```python
Z = 1
n = 3
l = 1
energy = -(Z**2) / (n**2)

dx = 1e-3
xmin = dx
xmax = 80

direction = "backward"
epsilon_u = 1e-10
```

```python
x_list, u_list = integrate_hydrogen(energy, Z, n, l, dx, xmin, xmax, direction, epsilon_u)
```

```python
from sympy.abc import r as sympy_r
from sympy.physics import hydrogen
from sympy.utilities.lambdify import lambdify

R_nl = lambdify(sympy_r, hydrogen.R_nl(n, l, sympy_r, Z))
u_list_symyp = R_nl(x_list) * x_list
```

```python
import matplotlib.pyplot as plt

plt.plot(x_list, u_list, label="Numerov")
plt.plot(x_list, u_list_symyp, "C1--", label="Sympy")
plt.legend()

plt.show()
```
